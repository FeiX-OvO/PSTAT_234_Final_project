## 1. Introduction 

News recommendation is an important technique for personalized news service. Movie recommendations and product recommendations have been well studied, however news recommendations have not due to lack of high-quality datasets. In this project, we aim to explore the news recommendation models: NRMS, NAML, and LSTUR. We use the MIND dataset, that is a large-scale dataset for news recommendation. Constructed from user click logs of Microsoft News, the dataset contains 1 million users and over 160k English news articles, each with a rich textual content such as title, abstract and body [1].

In traditional recommender systems, users and items are usually represented using IDs, and their interactions such as rating scores are used to learn ID representations. However, news recommendation poses special challenges. For instance, news articles are posted continuously, news sites update quickly, and articles are quickly removed from the news' main page. It is not sufficient to simply represent news articles by their IDs since understating the text is important. Also, there are no clear-cut ratings in news (as we do not typically rate news articles), so users’ interest in news is usually inferred from their click behaviors in an implicit way. News recommendation aims to find news articles that users have interest to read from a massive amount of candidate news. Thus, the two main problems in News Recommendation are (1) how to represent news articles which have rich textual content and (2) How to model users’ interest in news from their previous behaviors.

#### The MIND Dataset
The MIcrosoft News Dataset (MIND) has helped facilitate research in this area. It was collected from the user behavior to logs of Microsoft News. The MIND dataset contains 1 million randomly sampled users who had at least 5 news click records during 6 weeks from October 12 to November 22, 2019. The behavior logs of these users in this period were collected, which are formatted into impression logs. An impression log records the news articles displayed to a user when she visits the news website homepage at a specific time, and her click behaviors on these news articles. Since in news recommendation we usually predict whether a user will click a candidate news.


This dataset is a large-scale dataset for news recommendation research.It was collected by Microsoft from Microsoft News website. It contains about 160k English news articles and more than 15 million impression logs generated by 1 million users. Every news article contains rich textual content including title, abstract, body, category and entities. Each impression log contains the click events, non-clicked events and historical news click behaviors of this user before this impression.
 
To protect user privacy, each user was de-linked from the production system when securely hashed into an anonymized ID.

```python
I = 16 # number of users
M = 15 # number of movies
K = 5  # number of latent factors

# initialize U and V with random values
np.random.seed(42)

U = np.random.uniform(0, 1, size=K*I).reshape((I, K))
V = np.random.uniform(0, 1, size=K*M).reshape((M, K))

Uold = np.zeros_like(U)
Vold = np.zeros_like(V)
```

```python
def rmse(X, Y):
    from numpy import sqrt, nanmean
    return sqrt(nanmean((X - Y)**2))

error = [(0, rmse(R, np.inner(V,U)))]

# calculate maximum magnitude of relative updates
def max_update(X, Y):
    from numpy import inf
    from numpy.linalg import norm
    return norm(((X - Y)/Y).ravel(), inf)

update = [(0, max(max_update(Uold, U), max_update(Vold, V)))]
```

```python
rate = 0.1            # learning rate (step size) 
max_iterations = 1500  # maximum number of iterations
threshold = 0.001     # max_update threshold for termination

for t in range(1, max_iterations):
     
    for m, i in zip(*np.where(~np.isnan(R))):
        
        U[i] = U[i] + rate*V[m]*(R.iloc[m,i] - np.inner(V[m], U[i]))
        V[m] = V[m] + rate*U[i]*(R.iloc[m,i] - np.inner(V[m], U[i]))
        
    # compute error after one sweep of updates
    error += [(t, rmse(R, np.inner(V,U)))]
    
    # keep track of how much U and V changes
    update += [(t, max(max_update(Uold, U), max_update(Vold, V)))]
    Uold = U.copy()
    Vold = V.copy()
    
error = pd.DataFrame(error, columns=['iteration', 'rmse'])
update = pd.DataFrame(update , columns=['iteration', 'maximum update'])
```

## 3. News Recommendation Models
We explore three different news recommendation models in this project.

### 3.1. NAML Model
The Neural News Recommendation with Attentive Multi-View Learning (NAML) model uses a neural news recommendation approach which can learn informative representations of users and news by exploiting different kinds of news information. The core of the NAML approach is a news encoder and a user encoder. In the news encoder they propose an attentive multi-view learning model to learn unified news representations from titles, bodies and topic categories by regarding them as different views of news [2].

The NAML model incorporates different kinds of news article information, such as title, body and topic category, processes different kinds of news information differently, and accounts for the fact that different news information may have different informativeness for different news (i.e., some news titles are longer and specific while others are short and vague). 

The core components are a news encoder and a user encoder. In the news encoder, they propose an attentive multi-view learning framework to learn unified news representations from titles, bodies and categories by incorporating them as different views of news. In addition, since different words and views may have different informativeness for news representation, they apply word-level and view-level attention networks to select important words and views for learning informative news representations. In the user encoder, they learn the representations of users from the representations of their clicked news. Since different news may also have different informativeness for user representations, they apply attention mechanism to news encoder to select informative news for user representation learning. Extensive experiments on a real-world dataset show the approach can effectively improve the performance of news recommendation.

#### News Encoder

The news encoder module is used to learn representations of news from different kinds of news information, such as titles, bodies and topic categories. Since different kinds of news information have different characteristics, instead of simply merging them into a long text for news representation, the model uses an attentive multi-view learning framework to learn unified news representations by regarding each kind of news information as a specific view of news.

##### Title Encoder
The first layer is the title encoder, used to learn news representations from their titles. Denote the word sequence of a news title of length M as $[w^t_1,w^t_2, ..., w^t_M]$. The words are embedded into a sequence of word vectors $[e^t_1,e^t_2, ..., e^t_M]$ via a word embedding look-up table $\boldsymbol{W}_e \in R^{V \times D}$, where $V$ and $D$ are vocabulary size and word embedding dimension respectively. 
 
The second layer is a convolutional neural network (CNN). CNNs are used to learn contextual word representations by capturing their local contexts. For example, in the news title “Xbox One On Sale This Week”, the local contexts of “One” such as “Xbox” and “On Sale” are useful for understanding that it belongs to a game console name. The contextual representation of the i-th word $\boldsymbol{c}^t_i$ is calculated by: $$\boldsymbol{c}^t_i = ReLU(\boldsymbol{F}_t \times e^t_{(i-K):(i+K)} + \boldsymbol{b}_t)$$
 where $e^t_{(i-K):(i+K)}$ is the concatenation of word embeddings from position $(i − K)$ to $(i + K)$. $\boldsymbol{F}_t \in R^{N_f \times (2K + 1)D}$ and $\boldsymbol{b}_t \in R^{N_f}$ are the kernel and bias parameters of the CNN filters, $N_f$ is the number of CNN filters and $2K + 1$ is their window size. ReLU is a non-linear activation function. The output of this layer is a sequence of contextual word representations, i.e. $[c^t_1,c^t_2, ..., c^t_M]$. 
 
The third layer is a word-level attention network. The model uses a word-level attention network to select important words within the context of each news title. Different words in the same news title usually have different informativeness for learning news representations. For example, in the news title “Rockets End 2018 With A Win” the word “Rockets” is more informative than “With” in representing this news. The attention weight of the $i_{th}$ word in a news title is denoted by $\alpha^t_i$, which is formulated as: 
$$a^t_i = \boldsymbol{q}^T_t tanh(\boldsymbol{V}_t \times \boldsymbol{c}^t_i + \boldsymbol{v}_t)$$
$$\alpha_i^t = \frac{exp(a^t_i)}{\sum_{j=1}^{M} exp(a^t_j)}$$
where $\boldsymbol{V}_t$ and $\boldsymbol{v}_t$ are the projection parameters, $\boldsymbol{q}_t$ denotes the attention query vector.

The final representation of a news title is the summation of the contextual representations of it words weighted by their attention weights, i.e., $\boldsymbol{r}^t = \sum_{j=1}^{M} \alpha^t_j \boldsymbol{c}^t_j.$

##### Body Encoder
The second component in news encoder module is body encoder, which is used to learn news representations from their bodies. It is similar with the title encoder and there are also three layers in the body encoder. 

The first layer is the same as the first layer for the title encoder: word embedding. Now, the only differenc eis that the instead of looking at a length M title, we are now looking at a length P news article body. Denote the word sequence of a news body as $[w^t_1,w^t_2, ..., w^t_P]$. Through the word embedding layer, it is transformed into a sequence of word vectors $[e^t_1,e^t_2, ..., e^t_P]$. The second layer is a CNN network, which takes the word vector sequence as input, and learns the contextual word representations $[c^t_1,c^t_2, ..., c^t_P]$ by capturing local contexts. The third layer is an attention network applied to the body encoder, 

$$a^b_i = \boldsymbol{q}^T_b tanh(\boldsymbol{V}_b \times \boldsymbol{c}^b_i + \boldsymbol{v}_b)$$
$$\alpha_i^b = \frac{exp(a^b_i)}{\sum_{j=1}^{P} exp(a^b_j)}$$
 where $\boldsymbol{V}_b$ and $\boldsymbol{v}_b$ are the projection parameters, $\boldsymbol{q}_b$ denotes the attention query vector. The final represenation of a news title is the summation of the contexual represenations of it words weighted by their attention weights, i.e., $\boldsymbol{r}^b = \sum_{j=1}^{P} \alpha^b_j \boldsymbol{c}^b_j.$

##### Category Encoder
The third component in the news encoder module is category encoder, which is used to learn news representations from their categories. On many online news platforms such as MSN News, news articles are usually labeled with topic categories (e.g., “sports” and “entertainment”) and subcategories (e.g., “basketball NBA” and “gaming”) for targeting user interests which contain important information of these news. For example, if a user clicked many news articles with the category of “sports”, then we can infer that this user may be interested in sports news. The NAML model incorporates the category and subcategory information for news representations learning. The category encoder has inputs are the ID and category $v_c$ and the ID and subcategory $v_{sc}.$ The category encoder has two layers. 

The first layer is a category ID embedding layer. It converts the discrete IDs of categories and subcategories into low-dimensional dense representations (denoted as $e_c$ and $e_{sc}$ respectively). The second layer is a dense vector, used to learn the hidden category representations by transforming the category embeddings as: 

$$\boldsymbol{r}^c = ReLU(\boldsymbol{V}_c \times \boldsymbol{e}^c + \boldsymbol{v}_c)$$
$$\boldsymbol{r}^{sc} = ReLU(\boldsymbol{V}_s \times \boldsymbol{e}^{sc} + \boldsymbol{v}_s)$$
where $\boldsymbol{V}_c, \boldsymbol{v}_c, \boldsymbol{V}_s$ and $\boldsymbol{v_s}$ are parameters in dense layers.

##### Attentive Pooling 
The fourth component in the news encoder module is attentive pooling. Different kinds of news information may have different informativeness for learning the representations of different news. For example, the title "The
Astros improve outfield, agree to 2- year deal with Brantley" is more informative than the title "The best games of 2018." The first title should have a high weight in representing this news, however the second title is short and vague. The body and the topic category for the second title should have higher weights than the title for representing this news. Denote the attention weights of title, body, category and subcategory as $\alpha_t, \alpha_b, \alpha_c$ and $\alpha_{sc}$ respectively. The attention weight of the title view is calculated by: 
$$a^t = \boldsymbol{q}^T_v tanh(\boldsymbol{U}_v \times \boldsymbol{r}^t + \boldsymbol{u}_v)$$
$$\alpha_t = \frac{exp(a_t)}{exp(a_t) + exp(a_b) + exp(a_c) + exp(a_{sc})}$$ 

where $\boldsymbol{U}_v $ and $\boldsymbol{u}_v $ are projection parameters, $\boldsymbol{q}_v$ is the attention query vector. The attention weights of other news information such as body, category and subcategory can be computed in a similar way.

##### Final Unified Model
The final unified news representations learned by the news encoder module is the summation of the news representations from different views weighted by their attention weights: 
$$\boldsymbol{r} = \alpha_c\boldsymbol{r}^c + \alpha_{sc}\boldsymbol{r}^{sc} + \alpha_t\boldsymbol{r}^t  + \alpha_b\boldsymbol{r}^b $$

In the NAML approach the news encoder module is used to learn the representations of both historical news browsed by users and the candidate news to be recommended.

#### User Encoder
The user encoder module in our approach is used to learn the representations of users from the representations of their browsed news. Different news browsed by the same user have different informativeness for representing this user.
For example, the news “10 best NBA moments” is very informative for modeling user preferences since it is usually browsed by users who are interested in sports, but the news “The Weather Next Week” is less informative since it is clicked by massive users. The user encoder module applies a news attention network to learn more informative representations by selecting important news. The attention weight of the $i_{th}$ news browsed by a user is denoted as $\alpha^n_i$, calculated by: 

$$a^n_i = \boldsymbol{q}_n^T tanh(\boldsymbol{W}_n \times \boldsymbol{r}_i + \boldsymbol{b}_n)$$
$$\alpha_i^n = \frac{exp(a_i^n)}{\sum_{j=1}^{N} exp(a^n_j)}$$ 
where $\boldsymbol{q}_n, \boldsymbol{W}_n$, and $\boldsymbol{b}_n$ are the parameters in the news attention network and $N$ is the number of browsed news.

The final representations of news browsed by this user weighted by their attention weights: $$\boldsymbol{u} = \sum_{i=1}^{N} \alpha^n_i \boldsymbol{r}_i$$

#### Click Predictor
The click predictor module predicts the probability of a user browsing a candidate news based off their representations. Denote the representation of a candidate news $D^c$ as $r_c$ and the representation of user *u* as $\boldsymbol{u}$. The click probability score $\hat{y}$ is calculated by the inner product of the representation vectors of user *u* and the candidate news $D^c$, i.e., $\hat{y} = \boldsymbol{u}^T \boldsymbol{r}_c$.


### 3.2. LSTUR Model
The Neural News Recommendation with Long- and Short-term User Representations (LSTUR) model is like the NAML model, except the user encoder has a short-term user representation and a long-term user representation. The LSTUR model learns long-term user representations from the embeddings of their IDs and learns short-term user representations from their recently browsed news via gated recurrent (GRU) networks. The NAML model usually learn a single representation vector for each user and cannot distinguish the long-term preferences and short-term interests of users in reading news. Thus, the user representations learned in these methods may be insufficient for news recommendation. We focus on the differences between the NAML and the LSTUR user encoder. 

#### User Encoder 

##### Short-Term User Representation

Online users may have dynamic short-term interests in reading news articles, which may be influenced by specific contexts or temporal information demands. For example, if a user just reads a news article about “Mission: Impossible 6 – Fallout”, and she may want to know more about the actor “Tom Cruise” in this movie and click news articles related to “Tom Cruise”, although she is not his fan and may never read his news before. The LSTUR model proposes to learn short-term representations of users form their recent browsing history to capture their temporal interests and use a GRU network to capture the sequential news reading patterns. Denote news browsing sequence from a user sorted by timestamp in ascending order as $C = [c_1, c_2, ..., c_k]$, where $k$ is the length of the sequence. The news encoder is applied to obtain the representation of these browsed article, denoted at $[\boldsymbol{e}_1, \boldsymbol{e}_2, ..., \boldsymbol{e}_k]$. The short-term user representation is computed by: 

$$\boldsymbol{r}_t = \sigma(\boldsymbol{W}_r[\boldsymbol{h}_{t-1}, \boldsymbol{e}_t])$$
$$\boldsymbol{z}_t = \sigma(\boldsymbol{W}_z[\boldsymbol{h}_{t-1}, \boldsymbol{e}_t])$$
$$\tilde{\boldsymbol{h}_t} = tanh(\boldsymbol{W}_{\tilde{h}} [\boldsymbol{r}_t \odot \boldsymbol{h}_{t-1}, \boldsymbol{e_t}])$$
$$\boldsymbol{h}_t = \boldsymbol{z}_t \odot  \boldsymbol{h}_t + (1 -  \boldsymbol{z}_t) \odot \tilde{\boldsymbol{h}_t}   $$

where $\sigma $ is the sigmoid function, $\odot$ is the item wise product, $\boldsymbol{W}_r, \boldsymbol{W}_z$ and $\boldsymbol{W}_{\tilde{h}}$ are the parameters of the GRU network. The short-term user representation is the last hidden state of the GRU network, i.e., $\boldsymbol{u}_s = \boldsymbol{h}_k$.

##### Long-Term User Representations
Users also have long-term interests in reading news.  For example, a basketball fan may tend to browse many sports news related to NBA in several years. Thus, we propose to learn long-term representations of users to capture their consistent preferences. The LSUR model learns long-term user representations from the embeddings of their user IDs, which are randomly initialized and fine- tuned during model training. Denote *u* as the ID of a user and $\boldsymbol{W}_u$ as tje look-up table for long-term user representation, the long-term user representation of this user is $\boldsymbol{u}_l = \boldsymbol{W}_u [u]$


The LSTUR model then combines the long-term and short-term user presentations for one unified user representation by concatenating the long-term user representation with the short-term user representation as the final user representation (LSTUR-con).

### 3.3. NRMS Model

In this approach, the news representations are learnt by using multi-head self-attention to model the interactions between words. In the user encoder, the representations of users are learnt from their browsing history by using multi-head self-attention to capture their relatedness. Besides, additive attentions are applied to both news and user encoders to select important words and news to learn more informative news and user representations.

<center><img src="nrms.jpg"/></center>

#### News Encoder

The main difference from NAML model is that the second layer of news encoder in NRMS is a word-level multi-head self-attention network. The interactions between words are important for learning news representations. For example, in the news title “Rockets Ends 2018 with a Win”, the interaction between “Rockets” and “Win” is useful for understanding this news, and such long-distance interactions usually cannot be captured by CNN. In addition, a word may interact with multiple words in the same news. For instance, in above example the word “Rockets” has interactions with both “Ends” and “Win”. Thus, we propose to use multi-head self-attention to learn contextual representations of words by capturing their interactions. The representation of the $i$th word learned by the $k$th attention head is computed as $$\alpha^k_{i,j} = \frac{exp(\boldsymbol{e}_i^T \boldsymbol{Q}_k^w \boldsymbol{e}_j)}{\sum_{m=1}^M exp(\boldsymbol{e}_i^T \boldsymbol{Q}_k^w \boldsymbol{e}_m)},$$ $$\boldsymbol{h}^w_{i,k} = \boldsymbol{V}^w_k(\sum_{j=1}^M \alpha^k_{i,j}\boldsymbol{e}_j),$$ where $\boldsymbol{Q}_k^w$ and $\boldsymbol{V}^w_k$ are the projection parameters in the $k$th self-attention head, and $\alpha^k_{i,j}$ indicates the relative importance of the interaction between the $i$th and $j$th words. The multi-head representation $\boldsymbol{h}^w_{i}$ of the $i$th word is the concatenation of the representations produced by $h$ separate self-attention heads, i.e., $\boldsymbol{h}^w_{i} = [\boldsymbol{h}^w_{i,1}, \boldsymbol{h}^w_{i,2}, \dots, \boldsymbol{h}^w_{i,h}]$. The first layer is still word embedding and the third layer is an additive word attention network. The architecture can also be extended to multi-view news information. 

#### User Encoder

The user encoder module is used to learn the representations of users from their browsed news. It contains two layers. The first one is a news-level multi-head self-attention network. Usually, news articles browsed by the same user may have some relatedness. In addition, a news article may interact with multiple news articles browsed by the same user. Thus, multi-head self-attention is applied to enhance the representations of news by capturing their interactions. The representation of the $i$th news learned by the $k$th attention head is formulated as follows: $$\beta^k_{i,j} = \frac{exp(\boldsymbol{r}_i^T \boldsymbol{Q}_k^n \boldsymbol{r}_j)}{\sum_{m=1}^M exp(\boldsymbol{r}_i^T \boldsymbol{Q}_k^n \boldsymbol{r}_m)},$$ $$\boldsymbol{h}^n_{i,k} = \boldsymbol{V}^n_k(\sum_{j=1}^M \beta^k_{i,j}\boldsymbol{r}_j),$$ where $\boldsymbol{Q}_k^n$ and $\boldsymbol{V}^n_k$ are parameters of $k$th news self-attention head, and $\beta^k_{i,j}$ represents the relative importance of the interaction between the $j$th and the $k$th news. The multi-head representation of the $i$th news is the concatenation of the representations output by $h$ separate self-attention heads, i.e., $\boldsymbol{h}^n_{i} = [\boldsymbol{h}^n_{i,1}, \boldsymbol{h}^n_{i,2}, \dots, \boldsymbol{h}^n_{i,h}]$. The second layer is an additive news attention network like the NAML method.  

## 4. Experiments

### Model Training and Evaluation

We explore the full MIND dataset using NRMS and LSTUR model. They are both implemented by Microsoft research group using Tensorflow and Keras API. We use negative sampling techniques for model training. For each news browsed by a user (regarded as a positive sample), we randomly sample $K$ news which are shown in the same impression but not clicked by the user (regarded as negative samples). We shuffle the orders of these news to avoid possible positional biases. Denote the click probability score of the positive and the $K$ negative news as $\hat{y}^+$ and $[\hat{y}_1^-,\hat{y}_2^-, \dots, \hat{y}_K^-]$ respectively. These scores are normalized by the softmax function to compute the posterior click probability of a positive sample as follows: $$p_i=\frac{exp(\hat{y}_i^+)}{exp(\hat{y}_i^+)+\sum_{j=1}^K exp(\hat{y}_{i,j}^-)} $$ The news click probability prediction problem is re-formulated as a pseudo (K + 1)-way classification task, and the loss function for model training
is the negative log-likelihood of all positive samples $S$, which is formulated as follows: $$\mathcal{L} = -\sum_{i\in S}\log(p_i).$$

For validation, we input the user’s browsing history and a list of impression news. The predictions are then compared with the user’s behavior to evaluate the model performance. We can also rank the news in the impression by the click probability score. Notice that we do not select the news list for user, rather, we only rank them with the probability of clicking. We can use the metrics for classification problems to evaluate our results. First, the group AUC is calculated by taking the average of AUC scores for all the impressions in the validation set. AUC is equal to the probability that a classifier will rank a randomly chosen positive instance higher than a randomly chosen negative one, so a greater AUC means a better performance in classifying positive and negative labels. Since our result gives a ranking of the given list of news, both MRR (Mean Reciprocal Rank) and NDCG (Normalized Discounted Cumulative Gain) can be used to evaluate the ranked query. Hence mean MRR and NDCG for all impressions can be used to evaluate the overall model performance.


### Experiment Results

The group experimented on NSMR model and LSTUR model. Both models are from Microsoft Recommenders repository. The models are implemented with Tensorflow and Keras API. Following the definitions in the paper, the models are implemented with different user encoders and news encoders. The training model is `model` and the testing model is `scorer`. The most significant difference between them is that the training models include a `TimeDistributed` layer. This allows the model to apply the same layer onto data points in a time series. 

The training file we have are the news and behavior tables. For validation, we will input the user’s browsing history and a list of news. We will rank the given news by the probability of the user’s clicking. The predictions are then compared with the user’s behavior and some metrics are calculated. It’s important to notice that we are not selecting the news list for users, rather, we are only ranking them with the probability of clicking.

In the training process, entropy loss is used to calculate the loss and update the network. In the testing process, there are two methods: `fast_eval` function and `slow_eval` function. The `slow_eval` function uses the whole trained model to make predictions from `sigmoid` function, so the outputs are in the same format as the training results. The cons of the method is that it consumes much more time when doing the fitting end to end. The `fast_eval` function fits the user encoder and news encoder on the test files directly, then the results are combined using dot product. This was makes the evaluation pretty fast, but the outputs will not be bounded into 0 and 1, so we cannot calculate the cross entropy loss from the results. Instead, we can use other metrices to evaluate the results.

Since the performance of the user is either clicked or not, we can achieve this as a binary classification problem. For labels of the items are either ‘0’ or ‘1’.  Therefore, we can use the metrics for classification problems to evaluate our results. In this experiment, we are measuring the group AUC of the test set. Since our results will rank the given list of news, both mrr(mean reciprocal rank) and ndcg will be used to evaluate the results of the ranked query. 

The dataset we used is the full dataset from MIND dataset. The title words and abstract words are tokenized and converted into indexes. A pretrained word embedding matrix is generated from the existing words in training word dictionary and word vectors. The matrix is used as the initial weight of embedding layers. Besides the training and validation dataset, we are using a new test data to report the final results.

For NMRS model, the initial weights give a score of {'group_auc': 0.517, 'mean_mrr': 0.2221, 'ndcg@5': 0.2296, 'ndcg@10': 0.2914} and for LUSTR model, the initial weights give a score of {'group_auc': 0.519, 'mean_mrr': 0.2209, 'ndcg@5': 0.2286, 'ndcg@10': 0.2908}. The two initial scores are pretty similar. The LSTUR model is giving a higher score, one possible reason of which is that the model takes both words from titles and the abstracts while the NSMR model uses only the words from titles.

We used different training settings in the exploration of the model. First, we set epoch number to 50 and learning rate set to 0.0001. We got the loss versus epoch graph as the figure below. The loss decreases smoothly but is not converging with 50 epochs. When examining the gauc of the validation results, we see the gauc increases as the epoch ends. To achieve convergence of model, we increased learning rate to 1e-4. This time the training loss have a more obious trend of low learning rate. However, the gauc shows signs of decreasing after 10 epochs. At the end of 50 epochs, LSTUR has a much lower training loss compared with NRMS, but the gauc of both models are close.When we increase the learning rates to 3e-4, the training loss converges at the end of 50 epochs. However, LSTUR is having a better curve fitted to the data. In gauc graph, both models have a sharp decrease, while NRMS is more obvious than the other. The same trend also appears in the other metrices. After 10 epochs, all the evaluation scores decrease. This can be a sign of overfitting of the model.
